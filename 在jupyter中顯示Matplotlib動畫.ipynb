```python
%matplotlib notebook
import tensorflow as tf 
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import animation
fig, ax = plt.subplots()
x =np.arange(0,2*np.pi,0.01)

# 返回的是个列表
line, = ax.plot(x,np.sin(x))
def animate(i):
    # xdata 保持不变， ydata 更新成另外一批数据
    # 将0-100都传进去更新一下，i变化时，y也会变化，更新图像
    line.set_ydata(np.sin(x+i/10))
    return line,

def init():
    line.set_ydata(np.sin(x))
    return line,

# interval 是更新的频率，隔多少毫秒更新一次，这里是隔20ms更新一次
# blit=True,只更新有变化的点
ani = animation.FuncAnimation(fig=fig,func=animate,frames =100,init_func=init,interval =20,blit=False)
plt.show()
```

```python
%matplotlib notebook
import tensorflow as tf 
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import animation
import random
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"C:\Windows\Fonts\kaiu.ttf", size=10)  # C:\Windows\Fonts

fig, ax = plt.subplots()

x = random.sample(list(range(1,11)), 10)
y = random.sample(list(range(1,11)), 10)
point = np.stack((x, y), axis=-1)

# 返回的是个列表
line1, = ax.plot(x,y,color=(75/255, 139/255, 190/255, 1), label="path")
ax.scatter(x,y,color=(255/255, 212/255, 59/255, 1), label="point")

plt.legend(prop=font, loc=1)

plt.xlabel("x軸", fontproperties=font, size=20)
plt.ylabel("y軸", fontproperties=font, size=20)
def animate(i):
    # xdata 保持不变， ydata 更新成另外一批数据
    # 将0-100都传进去更新一下，i变化时，y也会变化，更新图像
    
    np.random.shuffle(point)  # 更新方法
    
    x, y = np.split(point, 2, axis=1)
    x = x.reshape(len(point))
    y = y.reshape(len(point))
    line1.set_xdata(x)
    line1.set_ydata(y)
    plt.title("epoch:%s"%i, size=20)

    return line1,

def init():
    line1.set_xdata(x)
    line1.set_ydata(y)

    return line1,

# interval 是更新的频率，隔多少毫秒更新一次，这里是隔20ms更新一次
# blit=True,只更新有变化的点
ani = animation.FuncAnimation(fig=fig,func=animate,frames =1000,init_func=init,interval =20,blit=False)
plt.show()
```

```python
%matplotlib notebook
import tensorflow as tf 
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import animation
import random
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"C:\Windows\Fonts\kaiu.ttf", size=10)  # C:\Windows\Fonts

fig, ax = plt.subplots()

x = random.sample(list(range(1,11)), 10)
y = random.sample(list(range(1,11)), 10)
point = np.stack((x, y), axis=-1)

# 返回的是个列表
line1, = ax.plot(x,y,color=(75/255, 139/255, 190/255, 0), label="path")
ax.bar(x,y,color=(255/255, 212/255, 59/255, 1), label="point")

plt.legend(prop=font, loc=1)

plt.xlabel("x軸", fontproperties=font, size=20)
plt.ylabel("y軸", fontproperties=font, size=20)
def animate(i):
    # xdata 保持不变， ydata 更新成另外一批数据
    y = [10]*10
    ax.bar(x,y,color=(75/255, 139/255, 190/255, 1), label="point")
    
    y = random.sample(list(range(1,11)), 10)  # 更新方法
 
    ax.bar(x,y,color=(255/255, 212/255, 59/255, 1), label="point")
    plt.title("epoch:%s"%i, size=20)

    return line1,

def init():
    line1.set_xdata(x)
    line1.set_ydata(y)

    return line1,

# interval 是更新的频率，隔多少毫秒更新一次，这里是隔20ms更新一次
# blit=True,只更新有变化的点
ani = animation.FuncAnimation(fig=fig,func=animate,frames =1000,init_func=init,interval =20,blit=False)
plt.show()
```
